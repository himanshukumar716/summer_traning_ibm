```python
import random

# ==========================================
# CRICKET MATCH REGISTERS (State Vectors)
# ==========================================
target_runs = 16  # Runs needed to win the super over
runs_scored = 0
wickets_down = 0
balls_bowled = 0

MAX_BALLS = 6     # 1 Over dynamic limit
MAX_WICKETS = 2   # A team gets maximum 2 wickets in this super over

print("=============================================================")
print("         WELCOME TO THE SUPER-OVER SIMULATOR       ")
print("=============================================================")
print(f"Target to Chase: {target_runs} runs in {MAX_BALLS} balls (1 Over).")
print(f"Batting Team Lineup: Maximum {MAX_WICKETS} wickets allowed.")
print("Rules: Choose your shot power (1 to 6) when facing the ball.")
print("Type '0' during the shot prompt to forfeit the match.")
print("=============================================================")

# Master Super-Over Loop
while balls_bowled < MAX_BALLS and wickets_down < MAX_WICKETS:
    
    print(f"\n--- BALL NUMBER: {balls_bowled + 1} ---")
    print(f"Scorecard -> Runs Scored: {runs_scored}/{wickets_down}")
    print(f"Equation  -> Need {target_runs - runs_scored} runs from {MAX_BALLS - balls_bowled} balls left.")
    
    # 1. Capture User Shot Choice
    user_input = input("Select Shot Power (1=Defend, 2=Placement, 4=Boundary, 6=Sixer) or 0 to Forfeit: ").strip()
    
    # 2. Handle Game Forfeit Check (Using Break Statement)
    if user_input == "0":
        print("\nMatch Forfeited! You walked off the field. Retired Hurt!")
        break
        
    # 3. Input Validation Constraint (Using Continue Statement)
    if user_input != "1" and user_input != "2" and user_input != "4" and user_input != "6":
        print("CRITICAL FAULT: Invalid batting stroke style selection!")
        print("Penalty: Bowler is running in again. Pick exactly 1, 2, 4, or 6.")
        continue  # Skips execution updates and restarts current ball
        
    shot_choice = int(user_input)
    balls_bowled += 1  # Legal delivery register updated
    
    # 4. Pseudorandom Bowler Delivery Engine
    # 1 = Dot Ball, 2 = Run-Saving Field, 3 = Wicket Delivery
    bowler_action = random.randint(1, 3)
    
    print("The Bowler runs in... delivers... and...")
    
    # ==========================================
    # COMBAT RESOLUTION MATRIX (Conditional Engine)
    # ==========================================
    
    # Case A: Bowler takes a Wicket!
    if bowler_action == 3:
        wickets_down += 1
        print(f"OUT!!! Clean bowled or caught in the deep! Wickets Down: {wickets_down}")
        
    # Case B: Excellent Bowling (Dot Ball or Great Fielding)
    elif bowler_action == 1:
        print("DOT BALL! Excellent delivery. The batsman played it straight to the fielder. No run!")
        
    # Case C: Runs Secured (Successful Shot)
    else:
        runs_scored += shot_choice
        if shot_choice == 6:
            print(f"IT'S A HUGE SIX!!! Clean hit over the stadium roof for 6 runs!")
        elif shot_choice == 4:
            print(f"FOUR RUNS! Traced across the grass like a rocket for a boundary!")
        else:
            print(f"Good running! Balanced connection adds {shot_choice} runs to the scorecard.")

    # 5. Real-Time Victory Threshold Check (Break Statement)
    if runs_scored >= target_runs:
        print("\nLIVE BREAKPOINT: Target successfully chased down!")
        break  # Snaps out of the loop instantly before completing the over

# ==========================================
# POST-MATCH PRESENTATION SUMMARY
# ==========================================
print("\n" + "="*45)
print("            POST-MATCH PRESENTATION          ")
print("="*45)
print(f"Final Team Score       : {runs_scored}/{wickets_down}")
print(f"Total Balls Delivered : {balls_bowled} out of {MAX_BALLS}")
print("-"*45)

if runs_scored >= target_runs:
    print(f"MATCH RESULT: YOU WON! Chased down the total with {MAX_BALLS - balls_bowled} balls remaining!")
elif wickets_down == MAX_WICKETS:
    print("MATCH RESULT: YOU LOST! All out! The bowling unit bundled you out.")
else:
    print(f"MATCH RESULT: YOU LOST! Defeat by {target_runs - runs_scored} runs. Out of balls!")
print("=============================================================\n")
```
